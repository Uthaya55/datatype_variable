# 2.create a Directed graph

```python
graph_mark=nx.DiGraph(Info="mark details")
```

# 3.Create Nodes along with positions

```python
graph_mark.add_node("mark",pos=(0,0))
graph_mark.add_node("student",pos=(3,8))
graph_mark.add_node("cat",pos=(0,-8))
graph_mark.add_node("animal",pos=(0,8))
graph_mark.add_node("CSU",pos=(5,0))
graph_mark.add_node("california",pos=(-5,0))
```

```python
pos=nx.get_node_attributes(graph_mark,"pos")
```

```python
graph_mark.add_edge("mark","cat",weight="has a")
graph_mark.add_edge("mark","student",weight="is a")
graph_mark.add_edge("mark","animal",weight="root")
graph_mark.add_edge("mark","california",weight="path")
graph_mark.add_edge("mark","CSU",weight="root")
```

```python
###### weight=nx.get_edge_attributes(graph_mark,"weight")
plt.figure()
nx.draw_networkx(graph_mark,pos,font_weight='bold',node_size=2000,font_size=7)
nx.draw_networkx_edge_labels(graph_mark,pos,edge_labels=weight)
```
