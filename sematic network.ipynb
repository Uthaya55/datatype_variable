```python
import networkx as nx
```

```python
import matplotlib.pyplot as mat
```

```python
%matplotlib notebook
```

```python
graph_UDHAY=nx.DiGraph(info="UDHAY from pmist")
```

```python
graph_UDHAY.add_node("UDHAY",pos=(-1,0))
graph_UDHAY.add_node("sonata",pos=(4,-5))
graph_UDHAY.add_node("kumbakonam",pos=(-6,5))
graph_UDHAY.add_node("thanjavur",pos=(4,5))
graph_UDHAY.add_node("cse",pos=(0,5))
graph_UDHAY.add_node("pmist",pos=(4,1))
graph_UDHAY.add_node("friend",pos=(-6,-1))
```

```python
pos=nx.get_node_attributes(graph_UDHAY,"pos")
```

```python
graph_UDHAY.add_edge("UDHAY","pmist",weight="study in")
graph_UDHAY.add_edge("UDHAY","kumbakonam",weight="lives")
graph_UDHAY.add_edge("friend","sonata",weight="giftes")
graph_UDHAY.add_edge("UDHAY","sonata",weight="wear")
graph_UDHAY.add_edge("UDHAY","cse",weight="studied")
graph_UDHAY.add_edge("UDHAY","pmist",weight="studied")
graph_UDHAY.add_edge("pmist","thanjavur",weight="located in")
weight=nx.get_edge_attributes(graph_UDHAY,"weight")
```

```python
mat.figure()
nx.draw_networkx(graph_UDHAY,pos,node_size=2000,font_weight="bold",font_size=7)
nx.draw_networkx_edge_labels(graph_UDHAY,pos,edge_labels=weight)
```
